### EDA for Binning Variables

#### WHY USE BINNING FOR FOLLOWING VARIABLES?
YR_AGE: Age is a continuous variable, which can help us better understand the default probability of different age groups.

AMT_LOAN_TOTAL: The total amount of loans is a continuous variable, which can help us understand the default risk of different loan amounts.

SCORE_SINGLE_DEBIT_CARD_LIMIT: This is a continuous variable that helps us understand the relationship between a borrower's credit score and the probability of default.

DAYS_BTW_APPLICATION_AND_FIRST_COLLECTION_OR_HIGH_RISK_CALL and DAYS_BTW_APPLICATION_AND_FIRST_COLLECTION_CALL: These are time series data, and breaking them down can help us understand the relationship between how soon after a loan is applied for a collection or a high-risk call is made and the risk of default.

AVG_LEN_COLLECTION_OR_HIGH_RISK_INBOUND_CALLS: This is a continuous variable that helps us understand the relationship between the average length of collections or high-risk calls and the risk of default.

Others: The above eda plots showed a unpredicdable patterns of these variables.

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Binning function
def binning(var):
    selected_data_df[var + '_bin'] = pd.qcut(selected_data_df[var], 10, duplicates='drop').values.add_categories("NoData")
    selected_data_df[var + '_bin'] = selected_data_df[var + '_bin'].fillna("NoData").astype(str)
    print(selected_data_df[var + '_bin'].value_counts(dropna=False))

features =['DAYS_BTW_APPLICATION_AND_FIRST_COLLECTION_CALL', 'SCORE_SINGLE_DEBIT_CARD_LIMIT', 
                  'YR_AGE', 'DAYS_BTW_APPLICATION_AND_FIRST_COLLECTION_OR_HIGH_RISK_CALL', 
                  'TD_CNT_QUERY_LAST_1MON_P2P', 'AVG_LEN_COLLECTION_OR_HIGH_RISK_INBOUND_CALLS', 'AMT_LOAN_TOTAL']


# Set the figure size
plt.figure(figsize=(10, 6))

for i in features:
    binning(i)

# Adding suffix _bin to new feature list
features_bin = [f + '_bin' for f in features]

# Perform EDA
for feature_bin in features_bin:
    # Calculate the count and mean for each bin
    z = selected_data_df.groupby(feature_bin)['loan_default'].agg(['count', 'mean']).reset_index()
    # Calculate the proportion for each bin
    z['count_percent'] = z['count'] / z['count'].sum()

    # Create subplot to show two metrics
    fig, ax1 = plt.subplots(figsize=(10, 6))

    # Plot the count proportion
    color = 'tab:red'
    ax1.set_xlabel('Bin')
    ax1.set_ylabel('Count Percent')
    ax1.bar(z[feature_bin], z['count_percent'])
    ax1.tick_params(axis='y')
    plt.xticks(rotation=90)  # Rotate x axis labels

    # Create the second y axis
    ax2 = ax1.twinx()
    color = 'tab:red'
    ax2.set_ylabel('Mean', color=color)
    ax2.plot(z[feature_bin], z['mean'], color=color)
    ax2.tick_params(axis='y', labelcolor=color)
    plt.title(f"Count Percent and Mean of loan_default by {feature_bin}")
    fig.tight_layout()
    plt.show()
```

### Conclusion for EDA Analysis

#### 1.Analysis of Skewed Variables:
In our data, the variables such as Age, TD_CNT_QUERY_LAST_1MON_P2P, TD_CNT_QUERY_LAST_1MON_SMALL_LOAN, TD_CNT_QUERY_LAST_3MON_P2P, TD_CNT_QUERY_LAST_3MON_SMALL_LOAN, and TD_CNT_QUERY_LAST_6MON_SMALL_LOAN present a noticeable normal distribution, all skewed to the left. This signifies that these variables show a left-skewness in their distribution, implying the majority of data are in a higher value range, with fewer values occupying the lower range. Specific to this case, this could signify that the majority of customers have a higher number of inquiries and age, with fewer customers possessing a lower number of inquiries and age.

#### 2.Detection of Outliers:
In the variables "DAYS_BTW_APPLICATION_AND_FIRST_COLLECTION_OR_HIGH_RISK_CALL" (the number of days between the application and the first collection or high-risk call) and "AVG_LEN_COLLECTION_OR_HIGH_RISK_INBOUND_CALLS" (average length of collection or high-risk inbound calls), we have detected outliers around -100 and -50.

#### 3.Comparison between Default and Non-Default Customers:
For the variables "MONTH_CREDIT_CARD_MOB_MAX" (Maximum months of credit card usage) and "SCORE_SINGLE_DEBIT_CARD_LIMIT" (Score of single debit card limit), their values are higher when the loan default is 0, that is, when there is no default. This could imply that customers with better credit status (longer credit card usage and higher single debit card limit) are more likely to repay loans on time.

#### 4.Analysis of Categorical Variable Characteristics:
For the categorical variables of Education Level (CODE_EDUCATION) and Total Loan Amount (AMT_LOAN_TOTAL), we observed that customers with lower education levels are more likely to default, and customers with smaller loan amounts also tend to default more easily. This could reflect that customers with higher education levels are more likely to repay loans on time, and customers with larger loan amounts are also more motivated to repay loans promptly due to the pressure of the loan amount.

## Section 3 Feature Engineering

```python
# Deal with the missing value
selected_data_df['DAYS_BTW_APPLICATION_AND_FIRST_COLLECTION_OR_HIGH_RISK_CALL'].fillna(selected_data_df['DAYS_BTW_APPLICATION_AND_FIRST_COLLECTION_OR_HIGH_RISK_CALL'].mean(), inplace=True)
selected_data_df['DAYS_BTW_APPLICATION_AND_FIRST_COLLECTION_CALL'].fillna(selected_data_df['DAYS_BTW_APPLICATION_AND_FIRST_COLLECTION_CALL'].mean(), inplace=True)
selected_data_df['AVG_LEN_COLLECTION_OR_HIGH_RISK_INBOUND_CALLS'].fillna(selected_data_df['AVG_LEN_COLLECTION_OR_HIGH_RISK_INBOUND_CALLS'].mean(), inplace=True)
```

```python
from sklearn.model_selection import train_test_split
from category_encoders import WOEEncoder

# select the features needed
feature_list = selected_data_df.columns[1:]

# split into train and test
X_train, X_test, y_train, y_test = train_test_split(selected_data_df[feature_list], selected_data_df['loan_default'], test_size=0.2, random_state=42)

# woe encoder
woe_encoder = WOEEncoder(cols=feature_list)

# use encoder to train
woe_encoder.fit(X_train, y_train)

# transform the train and test
X_train_woe = woe_encoder.transform(X_train)
X_test_woe = woe_encoder.transform(X_test)
```

```python
selected_data_df.info()
```

WOE (Weight of Evidence) encoding offers several benefits in Mortgage Probability of Default and Fraud analysis. It handles categorical variables by transforming them into continuous variables, preserving the monotonic relationship with the target variable. This enables the use of categorical features in models that require numerical inputs. WOE encoding captures non-linear relationships between categories and the likelihood of default or fraud, uncovering patterns that may exist in the data. It is less sensitive to outliers, making the encoding more robust and reliable. WOE encoding allows for easy comparison of predictive power across variables, aiding in the identification of influential features. The results are interpretable and actionable, providing clear insights into the strength and direction of the relationship between categories and the target variable. Overall, WOE encoding is a valuable tool in Mortgage Probability of Default and Fraud analysis, enhancing model performance, risk assessment, and decision-making processes.

## Section 4 Models

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import roc_auc_score, roc_curve
from sklearn.preprocessing import StandardScaler

# Standardize the features
scaler = StandardScaler()
X_train_woe_scaled = scaler.fit_transform(X_train_woe)
X_test_woe_scaled = scaler.transform(X_test_woe)

# logistic regression model
log_reg = LogisticRegression()
log_reg.fit(X_train_woe_scaled, y_train)

# Predict the probabilities
y_train_proba = log_reg.predict_proba(X_train_woe_scaled)[:,1]
y_test_proba = log_reg.predict_proba(X_test_woe_scaled)[:,1]

# Compute the ROC AUC
roc_auc_train = roc_auc_score(y_train, y_train_proba)
roc_auc_test = roc_auc_score(y_test, y_test_proba)

print(f'Train ROC AUC: {roc_auc_train}')
print(f'Test ROC AUC: {roc_auc_test}')
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve, auc

# Compute false positive rate and true positive rate
fpr_train, tpr_train, _ = roc_curve(y_train, y_train_proba)
fpr_test, tpr_test, _ = roc_curve(y_test, y_test_proba)

# Compute ROC AUC
roc_auc_train = auc(fpr_train, tpr_train)
roc_auc_test = auc(fpr_test, tpr_test)

# Initialize plot
plt.figure(figsize=(10, 6))

# Plot ROC curves
plt.plot(fpr_train, tpr_train, label=f'Train ROC curve (area = {roc_auc_train:.2f})')
plt.plot(fpr_test, tpr_test, label=f'Test ROC curve (area = {roc_auc_test:.2f})')

# Plot reference line
plt.plot([0, 1], [0, 1], color='navy', linestyle='--')

# Customize plot
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic')
plt.legend(loc="lower right")

plt.show()
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import roc_auc_score, roc_curve
import matplotlib.pyplot as plt

# Create a decision tree model and set the parameters
dt_model = DecisionTreeClassifier(max_depth=5, min_samples_split=18, min_samples_leaf=11, max_features='sqrt')

# Train the model
dt_model.fit(X_train_woe_scaled, y_train)

# Predict probabilities
y_train_proba_dt = dt_model.predict_proba(X_train_woe_scaled)[:, 1]
y_test_proba_dt = dt_model.predict_proba(X_test_woe_scaled)[:, 1]

# Calculate ROC AUC
roc_auc_train_dt = roc_auc_score(y_train, y_train_proba_dt)
roc_auc_test_dt = roc_auc_score(y_test, y_test_proba_dt)

print(f'Train ROC for Decision Tree: {roc_auc_train_dt}')
print(f'Test ROC for Decision Tree: {roc_auc_test_dt}')
```

Here, I set max_depth=5, meaning that the maximum depth of our decision tree is 5. We also set min_samples_split=18 and min_samples_leaf=11, meaning that the minimum number of samples required for node splitting is 18 and the minimum number of samples required for leaf nodes is 11. Finally, I set max_features='sqrt', meaning that the number of features to consider when finding the best sharding is the square root of the number of all features.

```python
# Compute the false positive rate (fpr), true positive rate (tpr), and thresholds for the ROC curve
fpr_train, tpr_train, thresholds_train = roc_curve(y_train, y_train_proba_dt)
fpr_test, tpr_test, thresholds_test = roc_curve(y_test, y_test_proba_dt)

# Plot the ROC curve
plt.plot(fpr_train, tpr_train, label='Train ROC Curve')
plt.plot(fpr_test, tpr_test, label='Test ROC Curve')
plt.plot([0, 1], [0, 1], 'k--', label='Random')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.legend()
plt.show()
```

Both Logistic Regression and Decision Tree models exhibit reasonable performance with training data, though they show signs of overfitting when applied to the test data. The Logistic Regression model outperforms the Decision Tree model on both datasets. This suggests that while both models are valid, Logistic Regression might provide better generalization for this particular problem. To improve these models, we could explore methods such as cross-validation, regularization, or ensembling, aiming to increase their robustness and capacity to generalize.
